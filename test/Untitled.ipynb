```python
import feedparser
import sys
import requests
import json
from newspaper import Article
import warnings
import pandas as pd

warnings.filterwarnings(action='ignore')

def get_news_contents(url):
    url = url
    article = Article(url, language='ko')
    article.download()
    article.parse()

    return article.text

def get_news(keyword:str='경제',count=3, contents=True,summary=False):
    
    CLIENT_ID = 'e2uxikffwu'
    CLIENT_SECRET = 'BaaYO0rGzAAfy0SpDeqlDHdhNFCaXZl7yYDvdEHp'
    

    
    google_news_rss_url = f"https://news.google.com/rss/search?q={keyword}&hl=ko&gl=KR&ceid=KR:ko"
    feed = feedparser.parse(google_news_rss_url)
    news_article = []
    
    news_frame = {
        'title' : [],
        'link' : [],
        'published' : []}
    
    cnt = 0
    for entry in feed.entries:
        if cnt < count:
            news_frame['title'].append(entry.title)
            news_frame['link'].append(entry.link)
            news_frame['published'].append(entry.published)
            
            cnt += 1
            
    if contents:
        for i in range(len(news_frame['title'])):
            url = str(news_frame['link'][i])
            news = get_news_contents(url)
            news_article.append(news)
        
        news_frame['contents'] = news_article
            
            
    if summary & contents:
        news_frame['summary'] = []
        url = 'https://naveropenapi.apigw.ntruss.com/text-summary/v1/summarize'

        headers = {
                'Accept': 'application/json;UTF-8',
                'Content-Type': 'application/json;UTF-8',
                'X-NCP-APIGW-API-KEY-ID': CLIENT_ID,
                'X-NCP-APIGW-API-KEY': CLIENT_SECRET
        }
        
        for i in range(len(news_frame['title'])):
            
            CONTENT = news_frame['contents'][i]
            

            
            data = {
                "document": {"content": CONTENT},"option": {"language": "ko", "model": "news", "tone": 2, "summaryCount": 4}}

            if len(CONTENT) < 20:
                news_frame['summary'].append(" ")
                pass
            else:
                response = requests.post(url, headers=headers, data=json.dumps(data).encode('UTF-8'), verify=False)
                rescode = response.status_code
                if(rescode == 200):
                    news_frame['summary'].append(json.loads(response.text)['summary'])
                else:
                    #print(response.text)
                    news_frame['summary'].append(' ')
        
    return pd.DataFrame(news_frame)
```

```python
import dash
from dash import html
from dash import dcc
import dash_bootstrap_components as dbc
from dash.dependencies import Input, Output, State
from flask import request
import pandas as pd

# 데이터
df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/solar.csv')

# 앱 선언
app = dash.Dash(__name__, external_stylesheets=[dbc.themes.BOOTSTRAP])

# 사이드바 레이아웃
SIDEBAR_STYLE = {
    "position": "fixed",
    "top": 0,
    "left": 0,
    "bottom": 0,
    "width": "16rem",
    "padding": "2rem 1rem",
    "background-color": "#f8f9fa",
}

sidebar = html.Div(
    [
        html.H2("Sidebar", className="display-4"),
        html.Hr(),
        html.P(
            "A simple sidebar layout with navigation links", 
            className="lead"
        ),
        dbc.Nav(
            [
                dbc.NavLink("Home", href="/", active="exact"),
                dbc.NavLink("Page 1", href="/page-1", active="exact"),
                dbc.NavLink("Page 2", href="/page-2", active="exact"),
            ],
            vertical=True,
            pills=True,
        ),
    ],
    style=SIDEBAR_STYLE,
)

# 메인 레이아웃
app.layout = dbc.Container(
    [
        dbc.Row([
            dbc.Col([
                sidebar,
                html.Button('Shutdown', id='shutdown-button')
            ], width=3),
            dbc.Col([
                dcc.Dropdown(id="dropdown", options=[{"label": x, "value": x} for x in df.State.unique()], value=df.State.unique()[0]),
                dcc.Input(id="input", placeholder="Enter something...", type='text'),
                html.Button(id='submit-button', n_clicks=0, children='Submit'),
                html.Div(id="output"),
                html.Hr(),
                dbc.Table.from_dataframe(df.head(5), striped=True, bordered=True, hover=True),
                html.Div(id='content')
            ], width=9)
        ]),
    ],
    fluid=True,
)

# 콜백
@app.callback(
    Output('content', 'children'),
    [Input('dropdown', 'value'), Input('submit-button', 'n_clicks')],
    [State('input', 'value')]
)
def update_output(value, n, text):
    if n > 0:
        return f'You\'ve entered: "{text}", and chosen: "{value}"'
    else:
        return f'You\'ve chosen: "{value}"'

@app.callback(
    Output('shutdown-button', 'n_clicks'),
    [Input('shutdown-button', 'n_clicks')]
)
def shutdown(n_clicks):
    if n_clicks:
        func = request.environ.get('werkzeug.server.shutdown')
        if func is None:
            raise RuntimeError('Not running with the Werkzeug Server')
        func()
        return 'Server shutting down...'
    return None

if __name__ == '__main__':
    app.run_server()
```
